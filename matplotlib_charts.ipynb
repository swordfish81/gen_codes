# Matplotlib charting subplots

```python
## Import packages. Define basic chart specs globally
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (18, 6)
plt.rcParams['font.size'] = '12'
plt.rcParams['hist.bins'] = '10'
```

```python
## Create some simple data using np
X = np.linspace(-np.pi, np.pi, 15)
theta = np.linspace(0,2*np.pi,15)
C = np.cos(X)
S = np.sin(X)
T = np.sin(theta)
CS = C-S

print(X)
print(C)
print(S)
print(T)
print(CS)
```

```python
## Plot axis
ax = plt.axes()

# 'bs:' mentions blue color, square 
# marker with dotted line.
ax1 = ax.plot(X, C, 'bs:', label='Cosine')

## Add some chart formating. Grid, legend, title
ax.set_title("This is a basic single plot. Just one timeseries")
plt.grid(axis='both')
plt.legend()
plt.show()
```

```python
## Make main axis
ax = plt.axes()

# 'bs:' mentions blue color, square 
# marker with dotted line.
ax1 = ax.plot(X, C, 'bs:', label='Cosine')
ax2 = ax.plot(X, S, 'ro-', label='Sin')

# Create new invisible x-axis from previous
bx=ax.twinx()

# Call new timeseries 
ax3 = bx.plot(X, CS*10, 'k--^', label='Diff of C, S *10')

# Label the two diff. y-scales
ax.set_ylabel('Scale for Cosine and Sin')
bx.set_ylabel('Scale for C-S*10')
ax.grid(axis='y', color='k', linestyle='--', linewidth=0.2)
bx.grid(axis='both')
# Add some chart formating. Grid, legend, title
ax.set_title("This is a dual single plot, sharing two timeseries with diff. y-axis scales")

#plt.legend()
plt.show()
```

```python
## Two independent charts, stacked side by side, not sharing any axes
fig, ax = plt.subplots(1, 2)
fig.suptitle('Subtitle plot 1x2', fontsize=18)

# Accessing each axes object to plot the data through returned array
ax[0].plot(X, C, 'bs:')
ax[0].set_ylabel('Making this common')
ax[0] = ax[0].twinx()
ax[0].plot(X, S*10, 'ro-')

ax[1].plot(X, T, 'k:^', label='Something')
ax[0].grid(axis='both')
ax[1].grid(axis='both')
ax[1].set_ylabel('Making this not common')
fig.tight_layout()

## Charts sharing y-axis, stacked on top of each other
fig1, bx = plt.subplots(2, 1, sharex=True)
fig1.suptitle('Subtitle plot 2x1', fontsize=18)
bx[0].plot(X, T, 'k:^')
bx[1].plot(X, C, 'b:')
bx[0].grid(axis='both')
bx[1].grid(axis='both')

bx[0].set_ylabel('Y-Label for bx[0]')
bx[1].set_ylabel('Y-Label for bx[1]')
bx[0].set_xlabel('For b[0] x-axis label')
bx[1].set_xlabel('For b[1] x-axis label')
fig.tight_layout()

## Just a standalone chart with dual axes
fig2 = plt.figure()

ax = fig2.add_axes([0.1, 0.1, 0.8, 0.8])
plt.title('This is the final plot', fontsize=18)
plt.grid(axis='both')
ax1 = ax.plot(X, C, 'bs:')
ax2 = ax.plot(X, S, 'ro-')
ax.set_ylabel('General Y-Label')
ax.set_xlabel('General X-Axis Label')
```

```python
## Two independent charts, stacked side by side, not sharing any axes
fig, ax = plt.subplots(2, 1, figsize=(18,9))
fig.suptitle('Subtitle plot 2x1', fontsize=18)
fig.tight_layout()

# Accessing each axes object to plot the data through returned array
ax[0].plot(X, C, 'bs:', label='Cosine')
ax[0].plot(X, S, 'ro-', label='Sin')
ax[0].grid(axis='both')
ax[0].set_ylabel('Left axis label')
ax[0].set_xlabel('This is the x-axis label')
ax[0] = ax[0].twinx()
ax[0].plot(X, CS*10, 'k--^', label='Diff of C, S *10')
ax[0].set_ylabel('Right axis label')
fig.tight_layout()

## Charts sharing y-axis, stacked on top of each other
fig1.suptitle('Subtitle plot', fontsize=18)
ax[1].plot(X, CS*10, 'k:')
ax[1].plot(X, CS*5, 'r-')
ax[1].grid(axis='both')
ax[1].set_ylabel('Left y-label for secondary plot')
ax[1].set_xlabel('X-axis for secondary plot')
fig.tight_layout()
```
