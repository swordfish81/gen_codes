```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (18, 6)
plt.rcParams['font.size'] = '12'
```

```python
X = np.linspace(-np.pi, np.pi, 15)
theta = np.linspace(0,2*np.pi,15)
C = np.cos(X)
S = np.sin(X)
T = np.sin(theta)
```

```python
print(X)
print(C)
print(S)
print(T)
```

```python
# [left, bottom, width, height]
bx = plt.axes([0.1, 0.1, 0.8, 0.8])
ax = plt.axes()

# 'bs:' mentions blue color, square 
# marker with dotted line.
ax1 = ax.plot(X, C, 'bs:')
  
#'ro-' mentions red color, circle 
# marker with solid line.
ax2 = ax.plot(X, S, 'ro-')
  
ax.legend(labels = ('Cosine Function', 'Sine Function'), loc = 'upper left')
  
ax.set_title("Trigonometric Functions")
plt.grid(axis='both')
plt.show()

ax3 = bx.plot(X, T, 'k:^')
bx.legend(labels = ('Theta Function'), loc = 'upper left')
  
bx.set_title("TNew Functions")
plt.grid(axis='both')
plt.show()
```

```python
## Two independent charts, stacked side by side, not sharing any axes
fig, ax = plt.subplots(1, 2)
fig.suptitle('Subtitle plot 1x2', fontsize=18)

# Accessing each axes object to plot the data through returned array
ax[0].plot(X, C, 'bs:')
ax[0].set_ylabel('Making this common')
ax[0] = ax[0].twinx()
ax[0].plot(X, S*10, 'ro-')

ax[1].plot(X, T, 'k:^', label='Something')
ax[0].grid(axis='both')
ax[1].grid(axis='both')
ax[1].set_ylabel('Making this not common')
fig.tight_layout()

## Charts sharing y-axis, stacked on top of each other
fig1, bx = plt.subplots(2, 1, sharex=True)
fig1.suptitle('Subtitle plot 2x1', fontsize=18)
bx[0].plot(X, T, 'k:^')
bx[1].plot(X, C, 'b:')
bx[0].grid(axis='both')
bx[1].grid(axis='both')

bx[0].set_ylabel('Y-Label for bx[0]')
bx[1].set_ylabel('Y-Label for bx[1]')
bx[0].set_xlabel('For b[0] x-axis label')
bx[1].set_xlabel('For b[1] x-axis label')
fig.tight_layout()

## Just a standalone chart with dual axes
fig2 = plt.figure()

ax = fig2.add_axes([0.1, 0.1, 0.8, 0.8])
plt.title('This is the final plot', fontsize=18)
plt.grid(axis='both')
ax1 = ax.plot(X, C, 'bs:')
ax2 = ax.plot(X, S, 'ro-')
ax.set_ylabel('General Y-Label')
ax.set_xlabel('General X-Axis Label')
```
